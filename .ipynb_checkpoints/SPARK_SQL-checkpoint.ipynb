```python
schema=StructType([
    StructField('sale_id',IntegerType()),
    StructField('product',StringType()),
    StructField('category',StringType()),
    StructField('sale_date',StringType()),
    StructField('region',StringType()),
    StructField('customer',StringType()),
    StructField('quantity',IntegerType()),
    StructField('unit_price',DoubleType()),
])
```

```python
df=spark.createDataFrame(sales_data,schema=schema)
```

```python
df.show(5)
```

```python
df.printSchema()
```

```python
df=df.withColumn('revenue',col('quantity')*col('unit_price'))
df.show(5)
```

# Implementing SPARK SQL

```python
df.createOrReplaceTempView('sales')
```

```python
spark.sql('SHOW TABLES').show()
```

```python
query1='''
        SELECT * FROM sales;
    '''

result=spark.sql(query1)
result.show()
```

```python
## Filter by category where category=Electronics

query2="""SELECT * FROM sales where category='Electronics'"""

result=spark.sql(query2)
result.show()
```

```python
## Group by product, order by revenue

query3="""
SELECT product,
    COUNT(*) AS orders,
    SUM(quantity) AS unit_sold,
    ROUND(SUM(revenue),2) AS total_revenue
FROM sales
GROUP BY product
ORdER BY total_revenue DESC
"""

result=spark.sql(query3)
result.show()
```

```python
#From the sales table, product details, region details, quantity details, unit_price details, revenue_details but revenue should be greater than 1000 and category should be electronics and revenue in desc

query4="""
SELECT product,region,quantity,unit_price,revenue FROM sales WHERE revenue>1000 AND category='Electronics'
ORDER BY revenue DESC
"""

result=spark.sql(query4)
result.show()
```

```python
## RANK  products by revenue within each region.

query5="""
SELECT region,product,
    ROUND(SUM(revenue),2) AS revenue,
    RANK() OVER (PARTITION BY region ORDER BY SUM(revenue) DESC) AS rnk
FROM sales
GROUP BY region,product
ORDER BY region,rnk
"""

result=spark.sql(query5)
result.show()
```

```python
#subquery=products over the average revenue

query6="""
SELECT product,category,total_revenue
FROM (SELECT product,category,SUM(revenue) AS total_revenue FROM sales GROUP BY product,category) prod_totals WHERE total_revenue>(SELECT AVG(revenue) FROM sales) ORDER BY total_revenue DESC
"""

result=spark.sql(query6)

result.show()
```
